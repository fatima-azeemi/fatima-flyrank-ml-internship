## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

import pandas as pd

# Load the starter dataset safely after directory setup
df = pd.read_csv('data/raw/content_refresh_anonymized.csv')

# Calculate 3 real numbers to back our lane
total_pages = len(df)
declining_pages = len(df[df['trend_direction'] == 'down'])
decline_pct = (declining_pages / total_pages) * 100
avg_word_count = df['word_count'].mean()

print(f"Number 1 - Total Pages in Slice: {total_pages:,}")
print(f"Number 2 - Pages in Decline (trend_direction == 'down'): {declining_pages:,} ({decline_pct:.1f}%)")
print(f"Number 3 - Average Word Count across pages: {avg_word_count:.0f} words")
```
